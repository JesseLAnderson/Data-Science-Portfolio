```python
vars1 = LpVariable.dicts("HTFRoutes", (hubs, focus), 0, cat="Continuous")
```

```python
vars2 = LpVariable.dicts("HTCRoutes", (hubs, cities), 0, cat="Continuous")
```

```python
vars3 = LpVariable.dicts("FTCRoutes", (focus, cities), 0, cat="Continuous")
```

```python
prob = LpProblem("Shipping_Costs", LpMinimize)
```

```python
prob += (
    lpSum([htf[h][f] * vars1[h][f] for h in hubs for f in focus])+
    lpSum([htc[h][c] * vars2[h][c] for h in hubs for c in cities])+
    lpSum([ftc[f][c] * vars3[f][c] for f in focus for c in cities]),
    "Sum_of_Transporting_Costs",)
```

```python
for f in focus:
    prob+= (
    lpSum(vars1[h][f] for h in hubs) == lpSum(vars3[f][c] for c in cities),
    f"Flow_Control_for_{f}",)
```

```python
for f in focus:
    prob+= (
        lpSum(vars1[h][f] for h in hubs) <= focus_cap[f],
        f"Quantity_Into_Focus_City_{f}",
    )
```

```python
for c in cities:
    prob+= (
        (lpSum(vars2[h][c] for h in hub_cap)) + (lpSum(vars3[f][c] for f in focus_cap)) == cendem[c],
        f"Center_Demand_in_{c}",)
```

```python
for h in hubs:
    prob+= (
        (lpSum(vars1[h][f] for f in focus_cap)) + (lpSum(vars2[h][c] for c in cendem)) <= hub_cap[h],
        f"Hub_Capacity_in_{h}",)
```

```python
prob.writeLP
```

```python
prob.solve()
```

```python
print("Status:", LpStatus[prob.status])
```

```python
for var in prob.variables():
    print(f"{var.name} = {var.varValue}")
```
