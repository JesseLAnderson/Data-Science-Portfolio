```python
def drop_outliers_IQR(x):

   q1=x.quantile(0.25)

   q3=x.quantile(0.75)

   IQR=q3-q1

   not_outliers = x[~((x<(q1-1.5*IQR)) | (x>(q3+1.5*IQR)))]

   outliers_dropped = not_outliers.fillna(x.mean().reset_index())

   return outliers_dropped
```

```python
df_exp = drop_outliers_IQR(df_exp)
```
